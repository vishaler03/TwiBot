```python
def launch_twitter(): # launches twitter.com
    
    global driver
    
    driver=webdriver.Chrome(executable_path='C:/chromedriver.exe')
    
    driver.maximize_window()
    
    driver.get('https://www.twitter.com')
    
    print('Launching twitter...')
    
    time.sleep(5)
    
```

```python
def login_twitter(username,password): # logs you in to Twitter.com
    
    global driver
    
    print('Logging in...')
    
     username='youremail@email.com' #'Your username here'
    
     password='your password' #'Your password here'
    
    try:
        
        log_in=driver.find_element_by_xpath('//div/span/span[contains(text(),"Log in")]')
        
        log_in.click()
        
        user=driver.find_element_by_xpath('//input[@name="session[username_or_email]"]')
        
        pwd=driver.find_element_by_xpath('//input[@name="session[password]"]')
        
        user.send_keys(username)
        
        pwd.send_keys(password)
        
        time.sleep(random.randint(7,9))
        
        log=driver.find_element_by_xpath('//div/span/span[contains(text(),"Log in")]')
        
        log.click()
        
    except:
        
        print('Log in failed!')
        
        pass
```

```python
def get_url(): # dont modify this
    
    print('Visiting page...')
    
    global URL_of_profile
    
    global driver
    
    url=URL_of_profile
    
    try:
        
        driver.get(url)
        
        time.sleep(random.randint(7,9))
        
    except:
        
        pass
        
```

```python
def scroll(duration=10):
    
    print('Scrolling...')
    
    global driver
    
    try:
        
        body=driver.find_element_by_tag_name('body')
        
        for i in range(duration):
            
            body.send_keys(Keys.PAGE_DOWN)
            
            body.send_keys(Keys.PAGE_DOWN)
            
        body.send_keys(Keys.HOME)
        
        body.send_keys(Keys.HOME)
        
    except:
        
        pass        
        
```

```python
def like_last_five_posts():
    
    print('Liking posts...')
    
    global driver
    
    get_url()
    
    scroll()
    
    time.sleep(7)
    
    try:
        
        likes=driver.find_elements_by_xpath('//div')
        
        likes=[like for like in likes if like.get_attribute('aria-label')!=None and 'Likes' in like.get_attribute('aria-label') and 'Liked' not in like.get_attribute('aria-label')]
        
        for i in range(min(6,len(likes))): # 5 minimum posts to be liked
            
            try:
                
                likes[i].click()
                
                time.sleep(10)
                
            except:
                
                pass
            
        driver.find_element_by_tag_name('body').send_keys(Keys.HOME)
        
        driver.find_element_by_tag_name('body').send_keys(Keys.HOME)
        
        time.sleep(9)
        
    except:
        
        pass
        
```

```python
def send_messages(msg):
    
    print('Sending messages...')
    
    global driver
    
    global current_user
    
    driver.get('https://www.twitter.com')
    
    time.sleep(10)
    
    try:
        
        messages=driver.find_elements_by_xpath('//span[contains(text(),"Messages")]')
        
        messages[1].click()
        
        time.sleep(3)
        
        new_message=driver.find_elements_by_xpath('//a[@href="/messages/compose"]')
        
        new_message[0].click()
        
        time.sleep(3)
        
        search_people=driver.find_elements_by_xpath('//input[@aria-label="Search query"]')
        
        search_people[0].send_keys(current_user)
        
        time.sleep(7)
        
        list_of_users=driver.find_elements_by_xpath('//li[@role="listitem"]')
        
        if len(list_of_users)>0:
            
            if "messaged" not in list_of_users[0].text:
                
                list_of_users[0].click()
                
                time.sleep(7)
                
                next_=driver.find_elements_by_xpath('//span[contains(text(),"Next")]')
                
                next_[0].click()
                
                time.sleep(3)
                
                message_box=driver.find_elements_by_xpath('//div[@role="textbox"]')
                
                message_box[0].clear()
                
                time.sleep(3)
                
                n=len(msg)
        
                message_to_send=your_messages[random.randint(0,n-1)]
                
                message_box[0].send_keys(message_to_send)
                
                time.sleep(1)
                
                message_box[0].send_keys(Keys.ENTER)  
                
            else:
                
                driver.find_element_by_tag_name('body').send_keys(Keys.ESCAPE)
                
        else:
            
            driver.find_element_by_tag_name('body').send_keys(Keys.ESCAPE)

    except Exception as e:
        
        print(e)
        
        driver.find_element_by_tag_name('body').send_keys(Keys.ESCAPE)        
        
        time.sleep(3)
        
        driver.get('https://www.twitter.com')
        
        time.sleep(7)
        
```
